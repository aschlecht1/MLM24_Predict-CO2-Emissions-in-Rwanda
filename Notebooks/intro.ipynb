# Setup & Load Data #


```python
#Import
from pathlib import Path 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Create paths
data_dir = Path('../Data')
code_dir = Path('../Code')

```

```python
#Load Data
emission_data = pd.read_csv(
    data_dir / 'train.csv',
    index_col = 'ID_LAT_LON_YEAR_WEEK',)

#Print Data
emission_data
```

```python
#Group By Location
#https://pandas.pydata.org/pandas-docs/stable/reference/groupby.html#computations-descriptive-stats
#take note that some sites never report any emissions
emission_sites = emission_data.groupby(['latitude','longitude']).emission.agg(['mean','median', 'min', 'max','std','count'])
emission_sites
```

```python
#Write to Disk
emission_sites.to_csv(data_dir / 'emission_sites.csv')
```
